Вам необходимо собирать только те статьи, в которых встречается хотя бы одно требуемое ключевое слово. Эти слова определяем в начале кода в переменной, например:

Поиск вести по всей доступной preview-информации (это информация, доступная непосредственно с текущей страницы).

Дополнительная часть (необязательная)
Улучшить скрипт так, чтобы он анализировал не только preview-информацию статьи, но и весь текст статьи целиком.

Для этого потребуется получать страницы статей и искать по тексту внутри этой страницы.

Итоговый датафрейм формировать со столбцами: <дата> - <заголовок> - <ссылка> - <текст статьи>

```python
import requests
import datetime
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
%%time
def convert_to_pd_date(date_string):
    if len(re.findall(r'сегодня', date_string)) > 0:
           return pd.to_datetime(datetime.datetime.now().date())
    elif len(re.findall(r'вчера', date_string)) > 0:
           return pd.to_datetime(datetime.datetime.now().date() - datetime.timedelta(days=1))
    else:
            month = monthly_dict[re.findall(r'[а-я]{3,}', date_string)[0]]
            year = re.findall(r'\d{4}', date_string)[0]
            day = re.findall(r'\d{1,2}\s', date_string)[0].strip()
            return pd.to_datetime((year + month + day), format='%Y%m%d')

def get_text_from_the_url(url):
    result = requests.get(url)
    soup = BeautifulSoup(result.text, 'html.parser')
    text = soup.find('div', class_='post__text')
    return(text.get_text())
        
# Этот словарь нужен для конвертации в нормальный формат даты    
monthly_dict = {'января':'1','февраля':'2','марта':'3','апреля':'4','мая':'5','июня':'6','июля':'7','августа':'8','сентября':'9','октября':'10','ноября':'11','декабря':'12'}
KEYWORDS = ['python', 'sql']
df = pd.DataFrame()

# Выбираем, какое количество страниц хотим проанализировать
for i in range(1,2):
    
    urlstring = 'https://habr.com/ru/all/page' + str(i)
    result = requests.get(urlstring)
    soup = BeautifulSoup(result.text, 'html.parser')
    posts = soup.find_all('article', class_='post post_preview')
    for post in posts:
        article_dict = {}
        title = post.find('h2', class_='post__title').find('a', class_='post__title_link')
        article_dict['title_name'] = title.text
        article_dict['title_url'] = title.get('href')
        article_dict['date'] = convert_to_pd_date(post.find('span', class_='post__time').text)
#       Следующая строчка нужна, если хотим анализировать весь текст статьи  
        article_dict['text'] = get_text_from_the_url(article_dict['title_url'])
        for keyword in KEYWORDS:
            for element in article_dict.values():
                if keyword.lower() in str(element).lower():
                    article_dict['keyword'] = keyword
                    df = pd.concat([df, pd.DataFrame([article_dict])])
                    break
df.head(20)


```

В итоге должен формироваться датафрейм со столбцами: <почта> - <дата утечки> - <источник утечки> - <описание утечки>.


```python
EMAIL = ['ishegurov@1cbit.ru', 'ivan_shegurov@gmail.com', 'sky_1503@rambler.ru']
url = 'https://digibody.avast.com/v1/web/leaks'
df = pd.DataFrame()

for element in EMAIL:
    payload = {'email': element}
    query = requests.post(url, json=payload)
    if query.json()['status'] == 'ok' and query.json()['value']!= []:
        result = query.json()['value'][0]
#         print(result)
        row = {}
        row['email'] = result['username']
#         row['leak_date'] = result['leak_date']
#Сайт в каком-то странном формате отдаёт юникс тайм. Его нужно делить на 1000, выяснил эмпирически :)
        row['leak_date'] = datetime.datetime.fromtimestamp(result['leak_date']/1000).date()
        row['leak_source'] = result['leak_info']['title']
        row['description'] = result['leak_info']['description']
        df = pd.concat([df, pd.DataFrame([row])])

df.head()
```
